```python
from datasets import load_dataset, load_metric
import pandas as pd
import numpy as np
```

# STS_B

```python
# sentence1: sentence A
# sentence2: sentence B
# label: semantic relatedness gold score (on a 0-5 continuous scale)
```

```python
task = "stsb"
dataset = load_dataset("glue", task)

train = np.reshape([data[key] for data in dataset['train'] for key in data], (len(dataset['train']), len(dataset['train'].features)))
test = np.reshape([data[key] for data in dataset['test'] for key in data], (len(dataset['test']), len(dataset['test'].features)))
validation = np.reshape([data[key] for data in dataset['validation'] for key in data], (len(dataset['validation']), len(dataset['validation'].features)))

stsb_train = pd.DataFrame(train, columns = list(dataset['train'].features.keys()))
stsb_test = pd.DataFrame(test, columns = list(dataset['test'].features.keys()))
stsb_val = pd.DataFrame(validation, columns = list(dataset['validation'].features.keys()))

stsb_train
```

```python
stsb_train["label"] = stsb_train["label"].astype(float)
stsb_train[stsb_train.label < 0]
```

# SICK

```python
# pair_ID: sentence pair ID
# sentence_A: sentence A
# sentence_B: sentence B
# label: textual entailment gold label: entailment (0), neutral (1) or contradiction (2)
# relatedness_score: semantic relatedness gold score (on a 1-5 continuous scale)
# entailment_AB: entailment for the A-B order (A_neutral_B, A_entails_B, or A_contradicts_B)
# entailment_BA: entailment for the B-A order (B_neutral_A, B_entails_A, or B_contradicts_A)
# sentence_A_original: original sentence from which sentence A is derived
# sentence_B_original: original sentence from which sentence B is derived
# sentence_A_dataset: dataset from which the original sentence A was extracted (FLICKR vs. SEMEVAL)
# sentence_B_dataset: dataset from which the original sentence B was extracted (FLICKR vs. SEMEVAL)
```

```python
task = "sick"
dataset = load_dataset("sick", task)

train = np.reshape([data[key] for data in dataset['train'] for key in data], (len(dataset['train']), len(dataset['train'].features)))
test = np.reshape([data[key] for data in dataset['test'] for key in data], (len(dataset['test']), len(dataset['test'].features)))
validation = np.reshape([data[key] for data in dataset['validation'] for key in data], (len(dataset['validation']), len(dataset['validation'].features)))

sick_train = pd.DataFrame(train, columns = list(dataset['train'].features.keys()))
sick_test = pd.DataFrame(train, columns = list(dataset['test'].features.keys()))
sick_val = pd.DataFrame(train, columns = list(dataset['validation'].features.keys()))

sick_train
```
